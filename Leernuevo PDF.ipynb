# Guardamos en TXT para poder leer y tratar los datos como matriz

```python
import os
salida = open("prueba.txt","wb")

for pag in docum:
    texto =pag.get_text().encode("utf8")
    salida.write(texto)
salida.close()
```

```python
txt = open("prueba.txt","r",encoding="utf8")
cont= txt.readlines()
txt.close()
len(cont)
cont
```

# Eliminamos espacios en blanco

```python
noneElement=[]
for i in range(0,len(cont)):
    if cont[i] == " \n":
        noneElement.append(i)
a = len(noneElement)
for i in range(0,a):
    cont.remove(" \n")
print(len(cont))
```

```python
cont2=[]
for i in range(0,len(cont)):
    aux=cont[i].split(sep=' \n')
    cont2.append(aux[0])
cont2
```

```python
cont2[2]
```

# Eliminamos los signos de igual

```python
cont3 = []
for i in range(0,len(cont2)):
    aux=cont2[i].split(sep=' =')
    cont3.append(aux[0])
cont3
```

```python
for i in range(0,len(cont3)):
    print("Linea:{} texto: {}".format(i,cont3[i]))
```
